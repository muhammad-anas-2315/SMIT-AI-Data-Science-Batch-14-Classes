```python
html= int(input("Enter a html number:  "))
python = int(input("Enter a python number: "))
jawa = int(input("Enter a jawa number: "))
php = int(input("Enter a php number: "))
egg  = int(input("ENter a egg number :  "))
total = html+python+jawa+php+egg
percentage = total/500*10

if percentage>= 90:
    grade= "A+"
elif percentage>=80:
    grade = "A"
elif percentage>=70:
    grade = "B"
elif percentage >=60:
    grade = "C"
else:
    grade = "Fail"
    
print(f'''
       your result 

          Total : {total}
          percentage : {percentage}
          grade : {grade}''')
```
