# Continue aur Break

```python
for a in range(10):
    if a ==3:
        break
    else:
        print(a)
    
```

# continue

```python
for a in range(10):
    if a ==3:
        continue
    else:
        print(a)
    
```

```python
name = []
for a in range(20):
    user_name = input("Enter Your Name")
    if user_name == "admin":
        continue
    else:
        name.append(user_name)
print(name)
```

# while loop

```python
pas = "python123"
attemps = 3
while attemps > 0:
    user_input = input("enter password")
    if user_input == pas:
        print("access granted")
        break
    else:
        attemps -=1
        print("wrong password")

        
if attemps == 0:
    print("account blocked")
```
